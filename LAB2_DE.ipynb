PERCEPTRON FROM SCRACH

```python
# STEP 1: Given Data

# Inputs (marks)
X = [75, 65, 80]          # x1, x2, x3

# Weights
w = [4, 3, 5]             # w1, w2, w3

# Bias
b = -1000

# Learning rate
eta = 0.01

# Target output (Pass = 1, Fail = 0)
y = 1
```

```python
# STEP 2: Compute Net Input (z)
# z = w1*x1 + w2*x2 + w3*x3 + b

z = 0
for i in range(len(X)):
    z += w[i] * X[i]

z = z + b
print("Net input (z):", z)

```

```python
# STEP 3: Apply Activation Function


if z >= 0:
    y_pred = 1
else:
    y_pred = 0

print("Predicted Output:", y_pred)
```

```python
# STEP 4: Calculate Error

error = y - y_pred
print("Error:", error)
```

```python
# STEP 5: Weight Update (if needed)
# w_new = w_old + (eta * error * x)

if error != 0:
    for i in range(len(w)):
        w[i] = w[i] + (eta * error * X[i])

```

```python
# STEP 6: Bias Update
# b_new = b_old + (eta * error)

b = b + (eta * error)
```

```python
# STEP 7: Recalculate After Learning

z_new = 0
for i in range(len(X)):
    z_new += w[i] * X[i]

z_new = z_new + b

if z_new >= 0:
    y_pred_new = 1
else:
    y_pred_new = 0

```

```python
# STEP 8: Print Final Results
print("\nAfter Learning:")
print("Updated Weights:", w)
print("Updated Bias:", b)
print("New Net input (z):", z_new)
print("New Prediction:", y_pred_new)
```
